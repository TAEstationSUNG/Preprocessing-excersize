## rank() method — 동률 처리 방식
- `ascending=False` : 큰 값 → 1위 (예약 많은 호텔이 1등)
- `method='min'` : 동률이면 **공동 낮은 번호** → 1,1,3 (스포츠 순위표 방식)
- `method='first'` : 동률이면 **먼저 나온 행이 낮은 번호** → 1,2,3 (무조건 다른 번호)
- `method='average'` : 동률이면 **평균 순위** → 1.5,1.5,3

```python
# rsv_cnt_tb의 rsv_cnt컬럼을 예약 많은 순으로 순위를 매기고 그걸 새 컬럼으로 추가하는 코드
rsv_cnt_tb['rsv_cnt_rank']=rsv_cnt_tb['rsv_cnt'].rank(ascending=False, method='min')
                                        #ascending=False는 큰값에 낮은 순위번호
rsv_cnt_tb
```

## 여러테이블 합치기

```python
pd.merge(reserve_tb.query('people_num==1'),         #예약테이블에서 1인 예약 데이터만 필터링
         hotel_tb.query('is_business'),             # 호텔 테이블에서 비즈니스호텔만
         on='hotel_id', how='inner')            # on='hotel_id'는 두 테이블을 이 컬럼 기준으로 합친다는것
# how=inner 는 양쪽 다 있는 것만, 두 조건 모두 만족하는 데이터만 남아야한다는 것
```

```python
import gc

small_area_mst = hotel_tb \
    .groupby(['big_area_name', 'small_area_name'], as_index=False) \
    .size().reset_index()
# 호텔 테이블에서 big area + small area로 묶고 인덱스 말고 컬럼으로 만들고, 행 수를 세고,
# 인덱스를 해제해서 컬럼으로 꺼냄
print(small_area_mst)

small_area_mst.columns=['index','big_area_name','small_area_name','hotel_cnt']
# as_index=False + reset_index() 둘 다 써서 컬럼이 하나 더 생겼으니까 이름 부여

small_area_mst['join_area_id']= \
 np.where(small_area_mst['hotel_cnt']-1 >= 20,  # 호텔 수에서 자기 자신 뺴고 20개 이상이면 small,
          small_area_mst['small_area_name'],    # 아니면 big으로 넣어서 join_area_id 컬럼을 생성
          small_area_mst['big_area_name'])

small_area_mst.drop(['hotel_cnt','big_area_name'],axis=1,inplace=True)
# 재료로 쓴 hotel_cnt, big_area_name 컬럼 삭제하고 원본에 저장

base_hotel_mst=pd.merge(hotel_tb, small_area_mst, on='small_area_name') \
                .loc[:, ['hotel_id', 'join_area_id']]
# 호텔 테이블이랑 small_area_mst를 small_area_name 기준으로 합치고, 
# 전체 행에서 hotel_id랑 join_area_id 컬럼만 추출

del small_area_mst      # 중간에 만든 임시 테이블 삭제하고 메모리 정리
gc.collect()
```

## 추천 호텔 테이블 만들기

**목적**: 내가 예약한 호텔 근처의 다른 호텔을 추천해주는 테이블 생성

**"근처"의 기준**
- 소지역 호텔 수 >= 20 → 소지역 기준으로 추천
- 소지역 호텔 수 < 20  → 대지역 기준으로 추천 (호텔이 적으니까 범위를 넓힘)

**흐름**
1. 소지역/대지역 테이블을 join_area_id로 rename 후 concat (위아래로 쌓기)
2. hotel_id → rec_hotel_id로 rename (merge시 컬럼 구분용)
3. base_hotel_mst랑 join_area_id 기준으로 merge
4. 자기 자신 제외 (hotel_id != rec_hotel_id)

**최종 결과**
| hotel_id | rec_hotel_id |
|----------|-------------|
| H001     | H002        | ← H001 근처 호텔 추천
| H001     | H003        |

```python
# 예약한 호텔 근처에 있는 다른 호텔들을 추천하는 코드
#pd.concat은 위아래로 쌓는거라 컬럼 이름이 같아야함.

# 1단계 - 소지역/대지역 테이블을 위아래로 쌓기
recommend_hotel_mst=pd.concat([
    hotel_tb[['small_area_name', 'hotel_id']] \
    .rename(columns={'small_area_name': 'join_area_id'}, inplace=False), 

    hotel_tb[['big_area_name', 'hotel_id']] \
    .rename(columns={'big_area_name': 'join_area_id'}, inplace=False)
])  #concat은 위아래로 쌓는거라서 컬럼이름이 같아야하므로 둘다 join_area_id로 rename하고 쌓음

# 2단계 - 컬럼명 변경
recommend_hotel_mst.rename(columns={'hotel_id' : 'rec_hotel_id'}, inplace=True)
# 나중에 merge할때 hotel_id가 두 테이블에 다 있으면 헷갈리니까 구분해줌

# 3단계 - merge + 필요한 컬럼만 + 자기 자신 제외
pd.merge(base_hotel_mst, recommend_hotel_mst, on='join_area_id') \
    .loc[:, ['hotel_id', 'rec_hotel_id']] \
    .query('hotel_id != rec_hotel_id')
# join_area_id 기준으로 같은 지역 호텔끼리 연결하고 2개 컬럼만 추출. 자기 자신은 추천해서 제외
```

```python
# 고객별로 그룹지어서, 각 그룹을 예약일자 순으로 정렬하는 코드
# .apply()는 그룹을 통쨰로 받아서 가공
result = reserve_tb \
     .groupby('customer_id') \
        .apply(lambda group:        #위의 코드는 고객별 그룹을 생성하고, 각 그룹에 함수를 적용하는거.
            group.sort_values(by='reserve_datetime', axis=0, inplace=False),
            #group는 한고객의 예약 데이터 전체.
            include_groups=False)
#각 그룹에 함수를 적용.
#  group.sort_values(by='reserve_datetime', axis=0, inplace=False)는 reserve_datetime을 기준으로 오름차순 정렬하는 코드
# axis=0은 행 방향으로 정렬, inplace=False는 원본 데이터프레임을 변경하지 않고 정렬된 결과를 반환하는 코드

result
```

```python
# .shift()는 데이터를 아래로 밀어내는 함수. 이 코드는 2칸 아래로 밀기
result['before_price']= \
    pd.Series(result['total_price'].shift(periods=2))
# 2칸 밀렸으니까 위에 2개는 NaN
# shift 없이는 2번전 값을 현재 행에 붙일 방법이 없고, 실무에서 분석할때
# 2번 전 예약보다 이번에 더 비싼 호텔을 예약했고 구매력이 올라가고있다. 이런 변화를 볼 수 있음

result

# 위의 코드에서 apply로 고객별 정렬을 해서 인덱스가 뒤섞임. shift한 결과의 인덱스랑 result의 인덱스가 안맞을 수 있음
# pd.Series()로 감싸면 인덱스를 새로 리셋해서 result테이블이랑 인덱스를 맞춰줌
# pd.Series()는 인덱스가 붙어있는 1차원 배열. DateFrame의 컬럼 하나를 뽑으면 그게 Series임
```

## rolling()
창문(window)이 한 칸씩 내려가면서 창문 안의 값을 계산

**그냥 sum vs rolling sum**
- 그냥 sum → 전체 다 더함 (50,000+80,000+120,000+30,000+90,000 = 370,000)
- rolling sum → 3개씩 묶어서 더함

| 예약순 | total_price | 그냥 sum | rolling sum (window=3) |
|-------|-------------|---------|----------------------|
| 1번   | 50,000      | 370,000 | NaN  ← 3개 안참       |
| 2번   | 80,000      | 370,000 | NaN  ← 3개 안참       |
| 3번   | 120,000     | 370,000 | 250,000 ← 1+2+3번    |
| 4번   | 30,000      | 370,000 | 230,000 ← 2+3+4번    |
| 5번   | 90,000      | 370,000 | 240,000 ← 3+4+5번    |

- `window=3` : 창문 크기 (몇 개씩 묶을지)
- `min_periods=3` : 3개 미만이면 NaN
- `center=False` : 현재 행 포함 이전 값들로 계산

```python
# 과거 n건의 합계를 구하기. 
#고객별로 날짜순 정렬 후, 3번 연속예약 금액을 창문 sliding방식으로 합산하는 코드

# 고객별로 묶어서 날짜순으로 정렬
result=reserve_tb.groupby('customer_id') \
    .apply(lambda x: x.sort_values(by='reserve_datetime', ascending=True)) \
    .reset_index(drop=True) #이 코드는 인덱스를 새로 리셋


result['price_sum']=pd.Series(
    result.loc[:, ['customer_id', 'total_price']]  
    .groupby('customer_id')
    .rolling(center=False, window=3, min_periods=3).sum()
    #cunstomer, total 두 컬럼만 뽑고, 고객별로 묶어서(고객끼리 섞이면 안되니까) 3칸짜리 창문을 내려가면서 합계
    .reset_index(drop=True) #인덱스 새로 리셋
    .loc[:, 'total_price']
)

result
```

```python
# 이번 예약 전까지의 평균 금액을 구하는 코드. roling으로 이동평균 만들고 shift로 한칸 밀어서 현재 예약 제외
#고객별 날짜순 정렬
result = reserve_tb.groupby('customer_id') \
    .apply(lambda x: x.sort_values(by='reserve_datetime', ascending=True)) \
    .reset_index(drop=True) #인덱스 새로 리셋

result['price_avg'] = pd.Series(
    result
        .groupby('customer_id')
        ['total_price'].rolling(center=False, window=3, min_periods=1).mean()   #3개의 평균,min_periods=1은 NaN없이 처음부터 계산
        .reset_index(drop=True)
)

result['price_avg'] = \
    result.groupby('customer_id')['price_avg'].shift(periods=1) #shift(1)로 1칸 밀기

result
```

```python
import pandas.tseries.offsets as offsets
import operator

#이 예약 기준으로 과거 90일 안에 결제한 금액의 합계를 구하는 코드

# 1단계 -날짜 변환. 문자열을 날짜 타입으로 변환
reserve_tb['reserve_datetime'] = \
    pd.to_datetime(
        reserve_tb['reserve_datetime'],
        format='%Y-%m-%d %H:%M:%S'
    )

# 2단계 -자기 자신이랑 자기 자신을 merge
sum_table = pd.merge(
    reserve_tb[['reserve_id', 'customer_id', 'reserve_datetime']],
    reserve_tb[['customer_id', 'reserve_datetime', 'total_price']]
        .rename(columns={'reserve_datetime': 'reserve_datetime_before'}),
    on='customer_id' #같은 고객의 모든 예약끼리 연결하도록 customer_id기준으로 merge
)

# 3단계 -90일 조건 필터링
sum_table = sum_table[
    operator.and_(  #operator.and_ 는 두 조건을 동시에 만족하는 것만
        sum_table['reserve_datetime'] > sum_table['reserve_datetime_before'], #현재 예약이 과거 예약보다 나중이어야함(미래에서 과거 방향)
        sum_table['reserve_datetime'] + offsets.Day(-90)  #현재 날짜에서 90일 뺀 날짜(과거 예약 날짜), 즉 과거 90일 안에 있는 예약만
        <= sum_table['reserve_datetime_before']
    )
    # 4단계 -reserve_id별 합계
].groupby('reserve_id')['total_price'].sum().reset_index() #필터링된 과거 90일 예약들을 현재 예약 id별로 합산

sum_table.columns = ['reserve_id', 'total_price_sum']

# 5단계 -원본에 붙이고 NaN은 0으로
pd.merge(
    reserve_tb,
    sum_table,
    on='reserve_id',
    how='left'
).fillna(0)
```
